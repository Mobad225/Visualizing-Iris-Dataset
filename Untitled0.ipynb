# Histogram for each class

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.preprocessing import StandardScaler
from mpl_toolkits import mplot3d
from matplotlib.ticker import StrMethodFormatter
from sklearn.datasets import load_iris
dataset = load_iris()
X = dataset.data
y = dataset.target
dataframe = pd.DataFrame(X, y, columns = dataset.feature_names)
plt.style.use('seaborn-deep')
for i in range(0,3):
  print("For the %s Class" % dataset.target_names[i])
  for j in range(0,4):
    current = dataframe.iloc[i * 50 : (i + 1) * 50, j : j + 1].transpose()
    plt.hist(current, bins = 10, label = dataset.feature_names[j])
  plt.legend()
  ax = plt.gca()
  ax.set_xlabel('length (cm)')
  ax.set_ylabel('count')
  plt.show()
  plt.close()
  print()
```

# Plotting each 2 attributes together

```
for i in range(0, 4):
  for j in range(i + 1, 4):
    print()
    current = dataframe.iloc[:,[i,j]]
    for k in range(0, 3):
      a = current.iloc[k * 50 : (k + 1) * 50, :1]
      b = current.iloc[k * 50 : (k + 1) * 50, 1:]
      plt.scatter(a, b, label = dataset.target_names[k])
    plt.legend()
    ax = plt.gca()
    ax.set_xlabel(dataset.feature_names[i])
    ax.set_ylabel(dataset.feature_names[j])
    plt.show()
    plt.close()

```

# Plotting each 3 attributes together



```
for i in range(0, 4):
  for j in range(i + 1, 4):
    for k in range(j + 1, 4):
      print()
      ax = plt.gca(projection='3d')
      current = dataframe.iloc[:,[i,j,k]]
      for l in range(0, 3):
        a = current.iloc[l * 50 : (l + 1) * 50, :1]
        b = current.iloc[l * 50 : (l + 1) * 50, 1:2]
        c = current.iloc[l * 50 : (l + 1) * 50, 2:]
        ax.scatter(a, b, c, label = dataset.target_names[l])
      plt.legend()
      ax.xaxis.set_tick_params(labelsize = 8)
      ax.yaxis.set_tick_params(labelsize = 8)
      ax.zaxis.set_tick_params(labelsize = 8)
      ax.set_xlabel(dataset.feature_names[i], fontsize = 10)
      ax.set_ylabel(dataset.feature_names[j], fontsize = 10)
      ax.set_zlabel(dataset.feature_names[k], fontsize = 10)
      plt.show()
      plt.close()
```

# Correlation matrix between features + Visualization + Comment

```
corr = dataframe.corr()
print(corr.to_string(), end = '\n\n')
sn.heatmap(corr);
```

Comment : Diagonal values show that a feature is directly proportional to itself, and with a ratio of 100%.
features like petal width vs petal length, also seem to be directly proportional (as the correlation coefficient is +ve), and with a signifact ratio as well, so as one of them increases, the other is likely to increase with about the same value. On the other hand, features like sepal width vs petal length are obviously inversely proportional (as the correlation coefficient is -ve), and with about 40% ratio, so if one of them increases, the other is likely to decrease by 40% of that value.
Mainly : 

*   Petal length vs Petal width seem to be very related.
*   Petal length vs Sepal width seem to be inversely related.
*   Petal width vs Sepal width seem to be inversely related.
*   Sepal length vs Sepal width seem to be unrelated.
*   Sepal length vs Petal width or Petal length seem to be related.




# Extra : Dimensionality reduction on the iris dataset

```
cov = dataframe.cov()
df = dataframe.sub(dataframe.mean(axis=0), axis=1)
eigenvalues, eigenvectors = np.linalg.eig(cov)
ls = []
tot = 0
eigenvectors = eigenvectors.transpose()
for i in range(0, len(dataset.feature_names)):
  ls.append((eigenvalues[i], eigenvectors[i]))
  tot += eigenvalues[i]
ls.sort(reverse = True)
takenVal = 0
taken = []
for i in range(0, len(ls)):
  takenVal += ls[i][0]
  taken.append(ls[i][1])
  if takenVal/tot >= 0.95:
    break
W = np.array(taken)
well_positioned_mat = df.to_numpy()
reduced_data = pd.DataFrame(well_positioned_mat.dot(W.transpose()))
for i in range(0, 3):
  current = reduced_data.iloc[i * 50 : (i + 1) * 50, :]
  plt.scatter(current[0], current[1], label = dataset.target_names[i])
ax = plt.gca()
ax.set_xlabel('u1')
ax.set_ylabel('u2')
plt.legend()
plt.show()
plt.close()
print("Dimensions reduced from 4 to 2, Keeping an accuracy of %.2f%%!" %  (takenVal*100/tot) )
```
